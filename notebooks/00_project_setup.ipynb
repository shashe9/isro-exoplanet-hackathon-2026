Cell 2: Define repo + Drive paths

```python
from pathlib import Path
import os

GITHUB_USERNAME = "shashe9"
REPO_NAME = "isro-exoplanet-hackathon-2026"

# Colab local working clone path
REPO_PATH = Path(f"/content/{REPO_NAME}")

# Google Drive persistent cache path
DRIVE_ROOT = Path("/content/drive/MyDrive/exoplanet_hackathon")
DRIVE_DATA = DRIVE_ROOT / "data"
DRIVE_OUTPUTS = DRIVE_ROOT / "outputs"
DRIVE_MODELS = DRIVE_ROOT / "models"

print("Repo path:", REPO_PATH)
print("Drive root:", DRIVE_ROOT)
```

Cell 4: Create project structure inside the GitHub repo

```python
from pathlib import Path

repo_dirs = [
    "notebooks",
    "src",
    "src/data",
    "src/search",
    "src/features",
    "src/models",
    "src/utils",
    "data/sample",
    "outputs",
    "models",
    "docs"
]

for d in repo_dirs:
    (REPO_PATH / d).mkdir(parents=True, exist_ok=True)

print("GitHub repo structure created.")
for d in repo_dirs:
    print(REPO_PATH / d)
```

Cell 5: Create persistent Drive cache structure

```python
drive_dirs = [
    DRIVE_DATA / "raw_lightcurves",
    DRIVE_DATA / "processed_lightcurves",
    DRIVE_DATA / "metadata",
    DRIVE_DATA / "cache",
    DRIVE_OUTPUTS / "figures",
    DRIVE_OUTPUTS / "tls_results",
    DRIVE_OUTPUTS / "features",
    DRIVE_OUTPUTS / "logs",
    DRIVE_MODELS
]

for d in drive_dirs:
    d.mkdir(parents=True, exist_ok=True)

print("Drive cache structure created.")
for d in drive_dirs:
    print(d)
```

Cell 6: Write .gitignore

```python
gitignore_text = """
# Python
__pycache__/
*.pyc
*.pyo
*.pyd

# Jupyter
.ipynb_checkpoints/

# Virtual env / cache
.env
.venv
.cache/

# Large data and outputs
data/raw_lightcurves/
data/processed_lightcurves/
outputs/
models/

# OS / misc
.DS_Store
Thumbs.db
"""

gitignore_path = REPO_PATH / ".gitignore"
gitignore_path.write_text(gitignore_text.strip() + "\n")

print(".gitignore written to", gitignore_path)
print(gitignore_path.read_text())
```

Cell 7: Write requirements.txt

```python
requirements_text = """
numpy
pandas
matplotlib
scipy
astropy
lightkurve
transitleastsquares
wotan
astroquery
batman-package
scikit-learn
"""

requirements_path = REPO_PATH / "requirements.txt"
requirements_path.write_text(requirements_text.strip() + "\n")

print("requirements.txt written to", requirements_path)
print(requirements_path.read_text())
```

Cell 8: Create README placeholders so folders are visible in GitHub

```python
readme_targets = {
    REPO_PATH / "data" / "sample" / "README.md": "# Sample data\nSmall sample files only. Large data stays in Drive/local storage.\n",
    REPO_PATH / "outputs" / "README.md": "# Outputs\nStore only lightweight tracked outputs here if needed.\n",
    REPO_PATH / "models" / "README.md": "# Models\nLarge checkpoints should not be committed.\n",
    REPO_PATH / "docs" / "README.md": "# Docs\nNotes, report drafts, and experiment summaries.\n",
}

for path, text in readme_targets.items():
    path.write_text(text)

print("Placeholder READMEs created.")
```
